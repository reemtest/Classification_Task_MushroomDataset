## Mushroom Classification Project

### Problem Description
The goal of this project is to classify mushrooms as **edible (e)** or **poisonous (p)** based on 20 descriptive features such as cap shape, cap color, stem characteristics, habitat, and season.  
This is a **binary classification problem** where the target variable is the mushroom class (`edible` or `poisonous`).  

---

### Project Workflow

#### 1. Exploratory Data Analysis (EDA)
- Inspect the dataset structure (rows, columns, data types).

- Look for class imbalance in the target variable (edible vs poisonous).  

#### 2. Data Cleaning & Preprocessing
- Handle missing values or unknown entries if present.

- Stratify the dataset when splitting into **training and test sets** to maintain class balance.

#### 3. Feature Selection
To reduce dimensionality and improve generalization, apply feature selection techniques such as:  
- **Filter Methods:** Use statistical tests (e.g., Chi-square, ANOVA F-test, mutual information) to rank features.

1-Logistic Regression

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

results = []
all_model_predictions = {}

def evaluate_and_store(model_name, version, y_test, y_pred, y_proba):
    acc = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # Handle both decision_function (1D) and predict_proba (2D)
    if y_proba is not None:
        if y_proba.ndim == 1:  # decision_function output
            roc_auc = roc_auc_score(y_test, y_proba)
        elif y_proba.shape[1] == 2:  # predict_proba output
            roc_auc = roc_auc_score(y_test, y_proba[:, 1])
        else:  # fallback for unusual shapes
            roc_auc = roc_auc_score(y_test, y_proba.ravel())
    else:
        roc_auc = None

    results.append({
        "Model": model_name,
        "Version": version,
        "Accuracy": acc,
        "Precision": precision,
        "Recall": recall,
        "F1": f1,
        "ROC_AUC": roc_auc
    })

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

lr = LogisticRegression(max_iter=100, random_state=42)
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)
y_proba_lr = lr.predict_proba(X_test)

print("Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred_lr, target_names=["edible", "poisonous"]))

cm_lr = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues', xticklabels=["edible","poisonous"], yticklabels=["edible","poisonous"])
plt.title("Logistic Regression - Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

cv_scores = cross_val_score(lr, X_train, y_train, cv=5, scoring='accuracy')
print(f"5-Fold CV Accuracy Mean: {cv_scores.mean():.4f}")
print(f"5-Fold CV Accuracy Std: {cv_scores.std():.4f}")

evaluate_and_store("Logistic Regression", "Baseline", y_test, y_pred_lr, y_proba_lr)
all_model_predictions[("Logistic regression", "Baseline")] = (y_test, y_proba_lr)
```

1-Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
import pickle

param_grid_lr = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear']
}

grid_lr = GridSearchCV(LogisticRegression(max_iter=1000, random_state=42),
                       param_grid_lr, cv=5, scoring='accuracy', refit=True)
grid_lr.fit(X_train, y_train)

best_lr = grid_lr.best_estimator_
y_pred_lr = best_lr.predict(X_test)
y_proba_lr = best_lr.predict_proba(X_test)

print("Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred_lr, target_names=["edible","poisonous"]))

cm_lr = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues', xticklabels=["edible","poisonous"],
            yticklabels=["edible","poisonous"])
plt.title("Logistic Regression - Confusion Matrix")
plt.show()

cv_scores_lr = cross_val_score(best_lr, X_train, y_train, cv=5, scoring='accuracy')
print(f"5-Fold CV Accuracy Mean: {cv_scores_lr.mean():.4f}")
print(f"5-Fold CV Accuracy Std: {cv_scores_lr.std():.4f}")

evaluate_and_store("Logistic Regression", "Tuned", y_test, y_pred_lr, y_proba_lr)
all_model_predictions[("Logistic Regression", "Tuned")] = (y_test, y_proba_nb)

with open("best_logistic_regression.pkl", "wb") as f:
    pickle.dump(best_lr, f)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier


param_grid_dt = {
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10]
}

grid_dt = GridSearchCV(DecisionTreeClassifier(random_state=42),
                       param_grid_dt, cv=5, scoring='accuracy', refit=True)
grid_dt.fit(X_train, y_train)

best_dt = grid_dt.best_estimator_
y_pred_dt = best_dt.predict(X_test)
y_proba_dt = best_dt.predict_proba(X_test)

print("Decision Tree Classification Report:")
print(classification_report(y_test, y_pred_dt, target_names=["edible","poisonous"]))

cm_dt = confusion_matrix(y_test, y_pred_dt)
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Greens', xticklabels=["edible","poisonous"],
            yticklabels=["edible","poisonous"])
plt.title("Decision Tree - Confusion Matrix")
plt.show()

cv_scores_dt = cross_val_score(best_dt, X_train, y_train, cv=5, scoring='accuracy')
print(f"5-Fold CV Accuracy Mean: {cv_scores_dt.mean():.4f}")
print(f"5-Fold CV Accuracy Std: {cv_scores_dt.std():.4f}")

evaluate_and_store("Decision Tree", "Tuned", y_test, y_pred_dt, y_proba_dt)
all_model_predictions[("Decision Tree", "Tuned")] = (y_test, y_proba_nb)

with open("best_decision_tree.pkl", "wb") as f:
    pickle.dump(best_dt, f)
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5]
}

grid_rf = GridSearchCV(RandomForestClassifier(random_state=42),
                       param_grid_rf, cv=5, scoring='accuracy', refit=True)
grid_rf.fit(X_train, y_train)

best_rf = grid_rf.best_estimator_
y_pred_rf = best_rf.predict(X_test)
y_proba_rf = best_rf.predict_proba(X_test)

print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf, target_names=["edible","poisonous"]))

cm_rf = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Oranges', xticklabels=["edible","poisonous"],
            yticklabels=["edible","poisonous"])
plt.title("Random Forest - Confusion Matrix")
plt.show()

cv_scores_rf = cross_val_score(best_rf, X_train, y_train, cv=5, scoring='accuracy')
print(f"5-Fold CV Accuracy Mean: {cv_scores_rf.mean():.4f}")
print(f"5-Fold CV Accuracy Std: {cv_scores_rf.std():.4f}")

evaluate_and_store("Random Forest", "Tuned", y_test, y_pred_rf, y_proba_rf)
all_model_predictions[("Random Forest", "Tuned")] = (y_test, y_proba_nb)

with open("best_random_forest.pkl", "wb") as f:
    pickle.dump(best_rf, f)
```

4-SVM

```python
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pickle

# 1. Parameter distribution (linear only, small grid)
param_dist = {'C': [0.1, 1, 10], 'kernel': ['linear']}

# 2. Randomized search WITHOUT probability
rand_svm = RandomizedSearchCV(
    SVC(random_state=42, probability=False),  # No probability for speed
    param_distributions=param_dist,
    n_iter=2,       # very small for speed
    cv=3,
    scoring='accuracy',
    refit=True,
    random_state=42,
    verbose=1
)
rand_svm.fit(X_train, y_train)

# 3. Use the best SVM directly (no probability)
best_svm = rand_svm.best_estimator_

# 4. Predictions
y_pred_svm = best_svm.predict(X_test)

# 5. Reports
print("SVM Classification Report:")
print(classification_report(y_test, y_pred_svm, target_names=["edible", "poisonous"]))

cm_svm = confusion_matrix(y_test, y_pred_svm)
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Reds',
            xticklabels=["edible","poisonous"], yticklabels=["edible","poisonous"])
plt.title("SVM - Confusion Matrix")
plt.show()

# 6. Store results (use decision_function instead of predict_proba for ROC-AUC)
y_scores_svm = best_svm.decision_function(X_test)  # FAST
evaluate_and_store("SVM", "Tuned", y_test, y_pred_svm, y_scores_svm)
all_model_predictions[("SVM", "Tuned")] = (y_test, y_scores_svm)

# 7. Save model
with open("best_svm.pkl", "wb") as f:
    pickle.dump(best_svm, f)
```

#7. Model Comparison

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

for res, (y_test_model, y_proba_model) in all_model_predictions.items():
    # res is (model_name, version), e.g., ("SVM", "Tuned")
    model_name, version = res
    if y_proba_model.ndim > 1:
        y_scores = y_proba_model[:,1]
    else:
        y_scores = y_proba_model  # e.g., SVM decision_function

    fpr, tpr, _ = roc_curve(y_test, y_scores)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"{model_name} ({version}) AUC={roc_auc:.2f}")

plt.plot([0,1],[0,1],'k--')  # diagonal line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves of All Models")
plt.legend()
plt.show()
```
